```python
import matplotlib.pyplot as plt 
import numpy as np
import os 
```

```python
RESULTSDIR = "./results"
```

## Digits

```python
dataset = "digits"
save = False 

savedir = os.path.join(RESULTSDIR, dataset, "plots")
if not os.path.exists(savedir) and save:
    os.makedirs(savedir)

y = np.load(os.path.join("./data", f"{dataset}.npz"), allow_pickle=True)["y"]
for fname in os.listdir(os.path.join(RESULTSDIR, dataset)):
    if fname.split(".")[-1] != "npy":
        continue
    X_ = np.load(os.path.join(RESULTSDIR, dataset, fname))

    plotname = ".".join(fname.split(".")[:-1])
    fig, ax = plt.subplots()
    ax.scatter(X_[:,0], X_[:,1], c=y)
    ax.set_title(plotname)
    if save:
        plt.savefig(os.path.join(savedir, f"{plotname}.png"), dpi=250, format="png", bbox_to_inches=None)
    plt.show()
```
